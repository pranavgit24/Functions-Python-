Q1. What is the difference between a function and a method in Python?

A function is independent and is defined using the def keyword. It can be called on its own, not tied to any object. eg.

def greet(name):
    return f"Hello, {name}"

print(greet("Alice"))  # Output: Hello, Alice

A method is a function that belongs to an object (usually an instance of a class).It is called using dot notation: object.method(). eg.

name = "Alice"

print(name.upper())  # 'upper' is a string method; Output: ALICE

Q5. What are iterators in Python and how do they differ from iterables?

In Python, iterators and iterables are both used to loop through data. An iterable is any Python object capable of returning its members one at a time.Common examples: list, tuple, string, dict, set. It has a built-in __iter__() method that returns an iterator.

An iterator is an object that remembers the state as you iterate over it.

It implements two methods:

__iter__() – returns the iterator object itself.

__next__() – returns the next item, and raises StopIteration when done.
eg.

nums = [1, 2, 3]  #iterable
it = iter(nums)  # Convert iterable to iterator

print(next(it))  # 1
print(next(it))  # 2
print(next(it))  # 3
''' next(it) would raise StopIteration here '''

Q6. Explain the concept of Generators in Python and how they are defined.

Generators in Python are special kinds of iterators that allow us to iterate over data on the fly, producing values one at a time using the yield keyword. They are memory-efficient and great for working with large data sets or streams. It is defined like a function, but uses yield instead of return. When called, it does not run the code immediately but returns a generator object.
eg.

def count_up_to(max):
    num = 1
    while num <= max:
        yield num
        num += 1

counter = count_up_to(3)

print(next(counter))  # 1
print(next(counter))  # 2
print(next(counter))  # 3
'''' next(counter) would raise StopIteration '''


Q7. What are the advantages of using generators over using regular functions?

- Memory Efficiency
Generators do not store all values in memory, unlike lists or functions that return a full collection.

They yield one item at a time, so they can handle very large datasets efficiently.

- Lazy Evaluation
Generators produce values only when needed (on-demand).

This reduces CPU and memory usage for long or infinite sequences.

- State Persistence
They remember their state between successive calls using next().

No need for external state variables or loops in user code.

- Improved Performance
Especially useful in pipelines (e.g., reading and processing files line by line).

Reduces overhead of processing large or streaming data.

- Cleaner Code
yield makes iterative algorithms simpler and more readable than manually managing iterators or state.

eg. Regular function vs Generator function using yield

def get_numbers():
    return [i for i in range(5)]  #loads all at once

def generate_numbers():
    for i in range(1000000):
        yield i  #consumes very little memory for such a big iteration

Q8. What is a lambda function in Python and when is it typically used?

A lambda function in Python is a small, anonymous (unnamed) function defined using the lambda keyword instead of def. Syntax>> lambda arguments: expression. It can have any number of arguments, but only one expression.It automatically returns the result of the expression.It is meant for short, simple functions. Syntax>> lambda arguments: expression

Lambda functions are often used when we need a quick function, usually for a short time, especially with map(), filter(), reduce().
eg.

Q9. Explain the purpose and usage of the map() function in Python.

The map() function in Python is used to apply a function to every item in an iterable (like a list or tuple), returning a new iterable (a map object) with the results. The purpose is to transform or process a collection without writing a loop manually. Syntax>> map(function, iterable)
eg.

a = [1, 2, 3]
b = [4, 5, 6]
result = map(lambda x, y: x + y, a, b)  #applied over multiple iterables
print(list(result))  # Output: [5, 7, 9]




Q10. What is the difference between map(), reduce() and filter() functions in Python?

map():
Applies a given function to each item of an iterable (e.g., list, tuple) and returns an iterator that yields the results. It transforms each element individually.
filter():
Constructs an iterator from elements of an iterable for which a function returns true. It selects elements based on a condition.
reduce():
Applies a function cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value. It performs an aggregate computation.

eg.

from functools import reduce

numbers = [1, 2, 3, 4, 5]

''' map(): squares each number '''
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

''' filter(): selects even numbers '''
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]

''' reduce(): calculates the product of all numbers '''
product = reduce(lambda x, y: x * y, numbers, 1)
print(product)  # Output: 120

```python
#Q4. Write a python function that checks if a given number is prime or not from 1 to 200.

def func_prime(num):
    if num < 2  or num > 200:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
user_input = int(input("Enter a number: "))

result = func_prime(user_input)

if result:
    print(user_input, "is a prime number")
else:
    print(user_input, "is not a prime number")
```

```python
#Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self  # the iterator object itself

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

fib = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers

for num in fib:
    print(num, end=" ")
```

```python
#Q7. Implement a generator function that reads a file line by line and yields each line as a string.

def func_read_file(path):
    with open(path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Removes the newline character at the end

for line in func_read_file('example.txt'):
    print(line)
```

```python
#Q9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit

celsius_temps = [0, 20, 37, 100]

# Use map() with a lambda function to convert to Fahrenheit
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

# Print the result
print("Fahrenheit temperatures:", fahrenheit_temps)
```

```python
#Q10. Create a Python program that uses filter() to remove all the vowels from a given string

def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_string))

# Example usage
user_input = input("Enter a string: ")
result = remove_vowels(user_input)
print("String without vowels:", result)
```
